```python
import pandas as pd
```

```python
stats = pd.read_csv("CSVs/player_mvp_stats.csv")
```

```python
del stats["Unnamed: 0"]
```

```python
stats = stats.fillna(0)
```

```python
stats.to_csv("CSVs/pm_stats_refined.csv")
```

```python
predictors = ["Age", "G", "GS", "MP", "FG", "FGA", 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']
```

```python
train = stats[stats["Year"] < 2021]
test = stats[stats["Year"] == 2021]
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=.1)
```

```python
reg.fit(train[predictors], train["Share"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
```

```python
combo = pd.concat([test[["Player", "Share"]],predictions], axis=1)
```

```python
combo.sort_values("Share", ascending=False).head(10)
```

```python
combo = combo.sort_values("Share", ascending=False)
```

```python
combo["Rk"] = list(range(1, combo.shape[0]+1))
```

```python
combo = combo.sort_values("predictions", ascending=False)
combo["Predicted_Rk"] = list(range(1, combo.shape[0]+1))
```

```python
combo.head()
```

```python
# This is the top 5 PREDICTED NBA MVP Candidates
# So I basically need to do this for every year
```

```python
def add_ranks(predictions):
    predictions = predictions.sort_values("predictions", ascending=False)
    predictions["Predicted_Rk"] = list(range(1,predictions.shape[0]+1))
    predictions = predictions.sort_values("Share", ascending=False)
    predictions["Rk"] = list(range(1,predictions.shape[0]+1))
    predictions["Diff"] = (predictions["Rk"] - predictions["Predicted_Rk"])
    return predictions
```

```python
years = list(range(1991, 2022))
all_predictions = []
for year in years[5:]:
    train = stats[stats["Year"]<year]
    test = stats[stats["Year"]==year]
    reg.fit(train[predictors], train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    combination = add_ranks(combination)
    all_predictions.append(combination)
```

```python
# Now let's take the results of the other model
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=400, random_state=1, min_samples_split=5)
```

```python
# Let's improve the model a little bit using better predictors
```

```python
stats["NPos"] = stats["Pos"].astype("category").cat.codes
stats["NTm"] = stats["Tm"].astype("category").cat.codes
```

```python
rf_predictions = []
for year in years[5:]:
    train = stats[stats["Year"]<year]
    test = stats[stats["Year"]==year]
    rf.fit(train[predictors], train["Share"])
    predictions = rf.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    combination = add_ranks(combination)
    rf_predictions.append(combination)
```

```python
# Similar to the ridge results, lets put these in a csv
for i in range(len(rf_predictions)):
    rf_predictions[i]["Year"] = years[5:][i]

top5rf = []

for df in rf_predictions:
    top5 = df.head(5)
    top5rf.append(top5)

top5rf = pd.concat(top5rf, ignore_index=True)
top5rf.to_csv("results/rf_results.csv")
```
