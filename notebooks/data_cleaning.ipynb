```python
import pandas as pd
```

```python
mvps = pd.read_csv("C:/Prog/nbamvp/CSVs/mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[['Player', 'Year', 'Pts Won', 'Pts Max', 'Share']]
```

```python
mvps.head()
```

```python
players = pd.read_csv("C:/Prog/nbamvp/CSVs/players.csv")
```

```python
players.head()
```

```python
del players['Unnamed: 0']
del players['Rk']
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "Tot"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
```

```python
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head()
```

```python
# There is a multi level index now, so we drop 2
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
combined_data = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined_data.head()
```

```python
# The players who didnt score any points in the mvp voting will have NaN values in that column, so we replace the NaNs with zeros
combined_data[["Pts Won", "Pts Max", "Share"]] = combined_data[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined_data[combined_data["Pts Won"] == 0]
```

```python
teams = pd.read_csv("C:/Prog/nbamvp/CSVs/teams.csv")
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
teams.head(30)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
nicknames = {}

with open("C:/Prog/nbamvp/CSVs/nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined_data["Team"] = combined_data["Tm"].map(nicknames)
```

```python
stats = combined_data.merge(teams, how="outer", on=["Team", "Year"])
```

```python
del stats["Unnamed: 0"]
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```
